### **1)CORE SELECTION AND CORE LOSS CALCULATION**

**A) Core-Material-1**

```python
#Electrical Steel,  Grade EN10106, M235-35A, 0.35 mm thickness

B1=1.5 # T max
density1=7.6 # g/cm3

#In order to calculate Core loss at 500 Hz , Steinmetz equation is used
#Steinmetz equation, Pcore=k*(f^a)*(B^b)
#From datasheet of core
#We need to find "a" in steinmetz equaion by using linearity
Pcore_loss_50_material1=2.35 # W/kg at 50Hz at 1.5 T 
PCore_loss_400_material1=41.9 # W/kg at 400Hz at 1.5 T
a=log(PCore_loss_400_material1/ Pcore_loss_50_material1)/log(400.0/50.0)
Pcore_loss_500_material1=(500.0/400.0)**a*PCore_loss_400_material1
print 'Pcoreloss at 500 Hz for material-1=',Pcore_loss_500_material1,'W/kg'
```

[Core DataSheet ,Grade EN10106, M25-35A](http://perso.uclouvain.be/ernest.matagne/ELEC2311/T2006/NOFP.pdf)

**B) Core-Material-2**

```python
#ATI Grain Oriented Electrical Steel M-2

B2=1.5 # T max
density2=7.65 # g/cm3

#In order to calculate Core loss at 500 Hz , Steinmetz equation is used
#Steinmetz equation, Pcore=k*(f^a)*(B^b)
#From datasheet of core
#We need to find "a" in steinmetz equaion by using linearity
Pcore_loss_50_material2=0.668 # W/kg at 50Hz at 1.5 T 
PCore_loss_60_material2=0.9 # W/kg at 400Hz at 1.5 T
a=log(PCore_loss_60_material2/ Pcore_loss_50_material2)/log(60.0/50.0)
Pcore_loss_500_material2=(500.0/60.0)**a*PCore_loss_60_material2
print 'Pcoreloss at 500 Hz for material-2=',Pcore_loss_500_material2,'W/kg'
```

[Core DataSheet ,ATI Grain Oriented Electrical Steel M-2](https://www.atimetals.com/Documents/goes_technical_data2_v1.pdf)

### **2)TURN NUMBER AND CORE AREA DETERMANITION**

```python
K_lamination=0.96 # Stacking Factor
E1=3000.0 
# E1=4.44*N1*Ac*f*B*K_laminitation
# Let'S find N1 X A
N1Ac=E1/(4.44*500.0*1.5*0.96)
print'N1xAc=',N1Ac
#Let's say primary winding turn number:12
N1=18
N2=N1*100*105/100.0 # %5 extra turn for voltage drop on copper
Ac=N1Ac/N1
# square core dimension calculation ,x
x=sqrt(Ac*1000000)
print 'N1=',N1,'turns'
print 'N2=',N2,'turns'
print 'x=',x,'mm'
I2=6.5*10**6 / 300000.0
print 'I2=',I2,'A'
I1=2200  #Assuming losses
```

```python
#Skin Depth Calculation for Copper cable at 500 Hz
#p=1.68*10**-8 # Copper resistivity at 20 'C
c=0.00386 # Temperature coefficient of copper per degree
p=(1.68+ 90*c)*10**-8 # resistivity of copper at 110 'C
f=500.0 # Hz
w=2*pi*f
u=4*pi*0.999991e-7
Skin_Depth=sqrt((2*p) / (w*u))*1000 # in mm
print 'Skin_Depth=',Skin_Depth,'mm'

```

### **5)COPPER AND CORE LOSS CALCULATAIONS**

###### A) Copper Losses

By using AWG Table resistance values for AWG-3 and AWG-9, Primary Winding and Secondary Winding resistances are calculated


```python
#Primary winding copper resistance calculations:
R1_length=pi*d*N1
R1=R1_length*0.6465/1000.0/17 # divided by 17 means that 17 parallel cables
print 'R1=', R1,'mOhm'
```

```python
#Secondary  winding copper resistance calculations:
R2_length=pi*d*N2
R2=R2_length*2.599/1000.0 
print 'R2=', R2,'mOhm'
```

Copper Loss=Primary winding Copper Loss + Secondary Winding Copper Loss


```python
P_Copper_Loss_1=(R1*(I1**2))/1000
print 'P_Copper_Loss_1=',P_Copper_Loss_1, 'W'
P_Copper_Loss_2=(R2*(I2**2))/1000
print 'P_Copper_Loss_2=',P_Copper_Loss_2, 'W'
P_Copper_Loss_Total=P_Copper_Loss_1+P_Copper_Loss_2
print 'P_Copper_Loss_Total=',P_Copper_Loss_Total, 'W'

```

###### B) Core Losses

```python
Volume_Total=x*W*H
Volume_Window=x*L*Ww
Volume_Core=(Volume_Total - Volume_Window)/1000
print 'Volume_Core',Volume_Core,'cm^3'
```

```python
Core_Weight_1=Volume_Core*density1/1000
print 'Core_Weight_1=',Core_Weight_1,'kg'
Core_Weight_2=Volume_Core*density2/1000
print 'Core_Weight_2=',Core_Weight_2,'kg'
```

```python
Core_loss_1=Core_Weight_1*Pcore_loss_500_material1
print 'Core_loss_1=',Core_loss_1,'W'
Core_loss_2=Core_Weight_2*Pcore_loss_500_material2
print 'Core_loss_2=',Core_loss_2,'W'
```

*There are two different core losses defined above since I select two different core material in part 1.

*If we chose first material , core loss will be higher

*Therefore, We should chose material-2 in order to satisfy high efficiency.

### 6) EFFICIENCY CALCULATIONS

Efficiency=(Pout/(Pout+losses))*100

```python
Pout=6500000.0
P_loss1=P_Copper_Loss_Total+Core_loss_1
P_loss2=P_Copper_Loss_Total+Core_loss_2
Efficiency1=(Pout/(P_loss1+Pout))*100
Efficiency2=(Pout/(P_loss2+Pout))*100
print 'Efficiency1=',Efficiency1,'%'
print 'Efficiency2=',Efficiency2,'%'
```

** ATTENTION: Core-2 has higher efficiency than Core-1 since material-2 has lower core loss.This is very significant since power transformer should have high efficiency as soon as possible**
